```python
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\HP-PC\Desktop\ML model\ford_car_dataset.csv")
data.head(2)
data.describe()
```

```python
missing_val=data.isnull().sum()
print(missing_val)
dup_rows=data.duplicated().sum()
print(dup_rows)
print('Shape of the data before removing duplicate rows:',data.shape)
if dup_rows>0:
  data.drop_duplicates(inplace=True)
  print('duplicate rows removed')
data.shape
```

```python
#Independent features (X)
X=data.drop('price',axis=1)

#Dependent feature (Y)
Y=data['price']

print("Shape of dataset:-",data.shape)
print("Shape of X:-",X.shape)
print("Shape of Y:-",Y.shape)
```

```python
#Identifies the categorical columns in X.
categorical_columns=X.select_dtypes(include=['object']).columns
#Performd One-Hot encoding using pd.get_dummies().
X_encoded=pd.get_dummies(X,columns=categorical_columns,drop_first=True)
#converted the encoded data to int type
X_encoded=X_encoded.astype(int)
print(X_encoded.head())
print(X_encoded.columns)
```

```python
#Used StandardScaler from sklearn.preprocessing.
from sklearn.preprocessing import StandardScaler
#Numeric columns
numerical_columns=['year','mileage','tax','mpg','engineSize']
scaler=StandardScaler()
#Standard Scaling on numeric columns.
X_encoded[numerical_columns]=scaler.fit_transform(X_encoded[numerical_columns])
print(X_encoded.head())
```

```python
#Split the preprocessed data into training and testing sets using train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_encoded,Y,test_size=0.33,random_state=42)

print("Shape of X_train:-",X_train.shape)
print("Shape of X_test:-",X_test.shape)
print("Shape of y_train:-",y_train.shape)
print("Shape of y_test:-",y_test.shape)
```

```python
#Trains a Linear Regression model using sklearn.
from sklearn.linear_model import LinearRegression
model=LinearRegression()

#Fit the model on training data.
model.fit(X_train,y_train)

#Prints th intercept and coefficient of the model.
print("Intercept:",model.intercept_)
print("Coefficients:",model.coef_)
```

```python
y_pred=model.predict(X_test)
print("Predicted values:-")
print(y_pred[:10])
print("\nActual values:-")
print(y_test.head(10))
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("R2 score:",r2)

#----------Interpretation---------------
# If the R2 score is 0.83, it means that approximately 83% of the variation
# in car prices is explained by our model's features (such as mileage, year,
# engine size, tax, mpg, and car model). The remaining 17% is due to other
# factors not included in our dataset.

#----------Model Performance------------
# An R2 score above 0.80 indicates a strong, highly capable model for a
# real-world dataset like car prices. It means your predictions are overall very
# close to the actual values!
```

```python
import joblib
joblib.dump(model,'LR_model.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(X_encoded.columns,'columns.pkl')
```
